## least squares with features augmentation using regularized loss function

```python
DATA_TEST_PATH = '../data/test.csv' 
y_test, tX_test, ids_test = load_csv_data(DATA_TEST_PATH)

tX_test = np.delete(tX_test, [4,5,6,12,23,24,25,26,27,28], axis=1)
bias_test = np.ones(tX_test.shape[0])
tX_test = np.c_[bias_test.T, tX_test]
```

```python
lambdas = np.logspace(-5, 0, 15)
accuracy = 0.0
degree = 5 # so far, degree=5 gives the best result(tried 10, 15, 20, 30)
tX_aug = build_poly(tX, degree)
tX_test_aug = build_poly(tX_test, degree)
for ind, lambda_ in enumerate(lambdas):
    w_rr = ridge_regression(y, tX_aug, lambda_)
    y_pred = predict_labels(w_rr, tX_test_aug)
    acc = compute_accuracy(y_test, y_pred)
    if acc > accuracy:
        accuracy = acc
        best_lambda = lambda_
        print(acc)
w_rr = ridge_regression(y, tX_aug, best_lambda)
y_pred = predict_labels(w_rr, tX_test_aug)
acc = compute_accuracy(y_test, y_pred)
print(w_rr.shape)
print(acc)
```
